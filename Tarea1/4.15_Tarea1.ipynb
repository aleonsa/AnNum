Solución:

$B$ y $H$ están dados con sus valores constantes, por lo que debemos manejar a $Q$ respecto a dos variables $n$ y $S$. Usando la ecuación 4.27 del Chapra: 

$$ \Delta Q(\tilde{n},\tilde{S}) = \left | \frac{\partial Q}{\partial n} \right |\Delta \tilde{n} + \left | \frac{\partial Q}{\partial S} \right |\Delta \tilde{S} $$

$$ = \left | \frac{-1}{\tilde{n}^2} \frac{(BH)^{\frac{5}{3}}}{(B+2H)^{\frac{2}{3}}} \tilde{S}^{\frac{1}{2}} \right |\Delta \tilde{n} +  \left | \frac{1}{\tilde{n}} \frac{(BH)^{\frac{5}{3}}}{(B+2H)^{\frac{2}{3}}} \frac{1}{2 \tilde{S}^{\frac{1}{2}}} \right |\Delta \tilde{S}$$

Evaluando lo que está dentro de las barras de valor absoluto:

```python
import numpy as np
B = 20
H = 0.3
n = 0.03
S = 0.0003
dQn = np.abs((-1/(n**2)) * (B * H)**(5/3)/(B + 2*H)**(2/3) * S**(1/2))
print(dQn)
```

```python
dQS = np.abs((1/(n)) * (B * H)**(5/3)/(B + 2*H)**(2/3) * 1/(2*S**(1/2)))
print(dQS)
```


Estimaciones del error en $n$ y $S$:
$$ \Delta \tilde{n} =  0.003, \Delta \tilde{S} =  0.00003$$

Por lo tanto:
$$ \Delta Q(\tilde{n},\tilde{S}) =(50.7370)(0.003) + (2536.8535)(0.00003)$$



```python
dn = 0.003
dS = 0.00003

print(dQn*dn)
print(dQS*dS)
```

$$ \Delta Q(\tilde{n},\tilde{S}) =0.1522 + 0.0761$$




Esto se puede ver como:
$$\Delta Q(\tilde{n},\tilde{S}) = error\ debido\ a\ variacion\ en\ n + error\ debido\ a\ variacion\ en\ S$$

Se observa que la variación en $n$ es mayor, por lo que se debería intentar obtener una mejor medición de esta.

```python
dQ = dQn*dn + dQS*dS
print(dQ)
```


$$ \Delta Q(\tilde{n},\tilde{S}) \approx 0.2283$$
