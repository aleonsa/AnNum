Serie de McLaurin para el coseno:

$$ cosx = \sum_{n=0}^{\infty } \frac{(-1)^{n}}{(2n)!}x^{2n} $$

En la página 59 del Chapra podemos encontrar la explicación de que es conveniente relacionar los errores de aproximación con el número de sifras significaticas. Es posible demostrar (Scarborough, 1966) que si el siguiente criterio se cumple, se tendrá la seguridad que el resultado es correcto en al menos n cifras significativas:

$$ E_a < (0.5 \times 10^{2-n})% $$

para 8 cifras esto se traduce en:

$$ E_a < (0.5 \times 10^{-6})% $$

Donde $E_a$ es determinada usando:

$$ E_a = \frac{aproximacionActual-aproximacionAnterior}{aproximacionActual} $$

El cálculo puede ser llevado a cabo con el siguiente código escrito en Python: 

```python
import numpy as np
import math
#defininfmos la serie de mclaurin
def func(n,x):
    return ((((-1)**n)/(math.factorial(2*n)))*(x**(2*n)))

res = 1 #La primera aproximación es el primer término de la suma, que es 1
error = 0
i = 1
#iteramos

while(True):
    #se evalúa la serie en el término n = i
    resAnterior = res
    res = res + func(i,0.3*np.pi)
    error = (res-resAnterior)/(res)    
    print(f"Término {i}: {res}, error: {error}")
    if (abs(error)<0.000000005): 
        break
    i+=1 # aumentamos i
    
    
print(f"\nSe necesitaron {i} términos para aproximar cos x a 8 cifras")
```
