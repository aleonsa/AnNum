Para una ecuación cuadrática de la forma : $a x^{2}+bx+c = 0$ la <b>fórmula estándar</b> está dada por: 

$$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

Mientras que para minimizar la cancelación por resta en casos donde $b>>4ac$ se usa la <b>fórmula alternativa</b>:

$$ x = \frac{-2c}{b \pm \sqrt{b^2 - 4ac}} $$

Solución:

Sabiendo que:
$$a = 1, b = -5000.002, c = 10$$

y que las raíces exactas de la ecuación son: $$x_1 = 5000 , x_2 = 0.002$$
<hr>

Obteniendo $x_1$ con la fórmula estándar:

$$x_1^{*} = \frac{5000.002 + \sqrt{(-5000.002)^2 - 4(1)(10)}}{2(1)} $$
$$ = \frac{5000.002 + \sqrt{2.5 \times 10^{7} - 40}}{2} $$
$$ = \frac{5000.002+4999.9}{2} $$
$$ = \frac{9999.9}{2} $$
$$ = 4999.9 $$

Obteniendo $x_2$ con la fórmula estándar:

$$x_2^{*} = \frac{5000.002 - \sqrt{(-5000.002)^2 - 4(1)(10)}}{2(1)} $$

$$ = \frac{5000.002 - \sqrt{2.5 \times 10^{7} - 40}}{2} $$

$$ = \frac{5000.002-4999.9}{2} $$

$$ = \frac{0.102}{2} $$

$$ = 0.051 $$

Calculando los errores relativos:

$$ \frac{\left | x_1 - x_1^{*} \right |}{\left | x_1 \right |}\times 100 = \frac{\left | 5000 - 4999.9 \right |}{\left | 5000 \right |}\times 100 = 0.002\%$$

$$ \frac{\left | x_2 - x_2^{*} \right |}{\left | x_2 \right |}\times 100 = \frac{\left | 0.002 - 0.051 \right |}{\left | 0.002 \right |}\times 100 = 24500\%$$

<hr>

Obteniendo $x_1$ con la fórmula alternativa:

$$x_1^{**} = \frac{-2(10)} {-5000.002 + \sqrt{(-5000.002)^2 - 4(1)(10)}}$$

$$ = \frac{-20} {-5000.002 + \sqrt{(2.5 \times 10^{7} - 40)}}$$

$$ = \frac{-20}{-5000.002+4999.9}$$

$$ = \frac{-20}{-0.102}$$

$$ 196.07 $$

Obteniendo $x_2$ con la fórmula alternativa:

$$x_2^{**} = \frac{-2(10)} {-5000.002 - \sqrt{(-5000.002)^2 - 4(1)(10)}}$$

$$ = \frac{-20} {-5000.002 - \sqrt{(2.5 \times 10^{7} - 40)}}$$

$$ = \frac{-20}{-5000.002-4999.9}$$

$$ = \frac{-20}{-9999.9}$$

$$ 0.002 $$

Calculando los errores relativos:

$$ \frac{\left | x_1 - x_1^{**} \right |}{\left | x_1 \right |} \times 100 = \frac{\left | 5000 - 196.07 \right |}{\left | 5000 \right |} \times 100 = 96.07 \%$$

$$ \frac{\left | x_2 - x_2^{**} \right |}{\left | x_2 \right |} \times 100 = \frac{\left | 0.002 - 0.002 \right |}{\left | 0.002 \right |} \times 100 = 0\%$$

La primera fórmula proporciona un cálculo más acertado de $x_1$ que para $x_2$ porque en el primer caso evita restar numeros similares. La segunda fórmula calcula $x_2$ de forma precisa, pero falla completamente en $x_1$. Por lo tanto cuando $b²>>4ac$ hay que calcular las raíces usando ambas fórmulas, dependiendo del signo de b habrá que evitar la resta de números parecidos.
